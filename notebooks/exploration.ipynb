# Exploration des données - Tickets Support

Ce notebook vous guide dans l'exploration du dataset de tickets support.

**Objectifs :**
- Comprendre la structure des données
- Analyser la distribution des catégories
- Explorer les caractéristiques textuelles
- Identifier les patterns utiles pour la classification

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuration
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

## 1. Chargement des données

```python
df = pd.read_csv('../data/sample_tickets.csv')
print(f'Shape: {df.shape}')
df.head(10)
```

## 2. Distribution des catégories

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Barplot
df['category'].value_counts().plot(kind='bar', ax=axes[0], color='steelblue')
axes[0].set_title('Distribution des catégories')
axes[0].set_ylabel('Nombre de tickets')

# Pieplot
df['category'].value_counts().plot(kind='pie', ax=axes[1], autopct='%1.1f%%')
axes[1].set_title('Proportion des catégories')

plt.tight_layout()
plt.show()
```
